```python
# %load prepro.py
import os
import pandas as pd
import re
import numpy as np 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# pd.set_option('display.width', 5000)
```

```python
TRAINING_PATH = './Dataset/Kaggle_Fake_News/train.csv'
TESTING_PATH = './Dataset/Kaggle_Fake_News/test.csv'
```

```python
column_name =["id", "title", "author", "text"]

trainX = pd.read_csv(TRAINING_PATH)

trainX.text = trainX.text.str.lower()
trainX.text = trainX.text.fillna(' ')
```

```python
stop = stopwords.words('english')

pat = r'\b(?:{})\b'.format('|'.join(stop))
#print(pat)
trainX['text'] = trainX['text'].str.replace(pat, '')
#trainX['text'] = trainX['text'].str.replace(r'\s+', ' ')
trainX['text'] = trainX['text'].str.replace(r'[^\w\s]+', '')
trainX['text'] = trainX['text'].str.replace('\d+' , '')
trainX['text'] = trainX['text'].str.lower()
```

```python
ps = PorterStemmer()

def stemSentence(sentence):
    token_words = sentence.split()
    stem_sentence = [ps.stem(token) for token in token_words]
    stem_sentence.append(" ")
    return ' '.join(stem_sentence)
```

```python
trainX['text'].isnull().sum()
```

```python
trainX['text'] = trainX['text'].apply(stemSentence)   # fucking fail

trainY = trainX.loc[:,["label"]]
```

```python
trainX['text'] = trainX.apply(lambda row : word_tokenize(row['text']) , axis = 1)
```
